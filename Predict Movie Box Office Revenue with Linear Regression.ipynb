```python
import pandas
from pandas import DataFrame     
import matplotlib.pyplot as plt #plt is the short name of matplotlib.pyplot

from sklearn.linear_model import LinearRegression #scikit is a machine learing module I used it for linear regression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv.csv')  #make object for analysis
```

```python
X = DataFrame(data, columns=['production_budget_usd']) #using data frame I am getting budget colum from data
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
#matplotlib.pyplot.scatter(X,y)
#matplotlib.pyplot.show()

plt.figure(figsize = (10,6))
plt.scatter(X,y, alpha=0.3) #alpha is used for make transparent
plt.title('Flim cost vs global revenue')
plt.xlabel('Production Budget $')
plt.ylabel('worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression = LinearRegression() #create object 
regression.fit(X, y)
```

```python
regression.coef_ #to check slope coefficient(theta_1)
#for each dollar it will show us revinue
```

```python
#intercept
regression.intercept_ #theta_0
```

```python
#matplotlib.pyplot.scatter(X,y)
#matplotlib.pyplot.show()

plt.figure(figsize = (10,6))
plt.scatter(X,y, alpha=0.3) #alpha is used for make transparent
plt.plot(X, regression.predict(X), color='red', linewidth=4)

plt.title('Flim cost vs global revenue')
plt.xlabel('Production Budget $')
plt.ylabel('worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
#Goodness of fit
regression.score(X, y) #film revinue % of production budget
```
